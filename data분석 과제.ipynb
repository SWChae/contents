조사 추가 과제 1

```python
xnum = np.array([jongro_people[1], jongro_people[1], jongro_people[4], jongro_people[4], jongro_people[7], jongro_people[7],
                jongro_people[1], jongro_people[1], jongro_people[1]])
```

```python
ynum = np.array([jongro_people[2], jongro_people[3], jongro_people[5], jongro_people[6], jongro_people[8], jongro_people[9],
                jongro_people[4], jongro_people[7], jongro_people[11]])
```

```python
a = {'ratio' : np.divide(ynum, xnum)}
```

```python
df = DataFrame(a, index=['남자비율', '여자비율', '한국인_남자비율', '한국인_여자비율', '외국인_남자비율', 
               '외국인_여자비율', '한국인비율', '외국인비율', '고령인구비율'])
```

```python
labels = '남자비율', '여자비율'
sizes = [df.loc['남자비율'], df.loc['여자비율']]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
```
